```python
import numpy as np
from sklearn.decomposition import KernelPCA
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv("twoSpirals.txt", sep="\t", header = None)
df.dropna()
X = df.iloc[:, 0:2]
y = df.iloc[:, 2:3]

T = 3

# Apply Kernel PCA with Gaussian kernel
kpca = KernelPCA(n_components=T, kernel='rbf')
X_kpca = kpca.fit_transform(X)

# Print the shape of the transformed data
print("Shape of transformed data:", X_kpca.shape)

X_train, X_test, y_train, y_test = train_test_split(X_kpca, y, test_size=0.2, random_state=42)

model2 = LinearRegression()
model2.fit(X_train, y_train)

y_pred_pca = model2.predict(X_test)
rmse_pca = np.sqrt(mean_squared_error(y_test, y_pred_pca))
print("RMSE for pca transformed data:", rmse_pca)

```
